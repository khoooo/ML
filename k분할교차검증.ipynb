# 1. 환경준비

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/boston.csv'
data = pd.read_csv(path)
data.head(3)
```

# 3. 데이터 준비

```python
target='medv'

x = data.drop(target, axis=1)
y = data[target]
```

```python
x = (x-x.min()) / (x.max()-x.min())
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,
                                                 random_state=2022)
```

# 4. 모델링

#### LinearRegression

```python
# 1. 불러오기
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

# 2. 선언하기
model = LinearRegression()

# 3. 학습하기, 예측하기, 평가하기
cv_score = cross_val_score(model,x_train,y_train,cv=10)


print(cv_score)
print('평균 :', cv_score.mean())
print('표준편차 :', cv_score.std())
```

#### KNN

```python
# 1. 불러오기
from sklearn.neighbors import KNeighborsRegressor     # y가 수치형이기 때문에
from sklearn.model_selection import cross_val_score   

# 2. 선언하기
model = KNeighborsRegressor()

# 3. 학습하기, 예측하기, 평가하기
cv_score = cross_val_score(model,x_train,y_train,cv=10)

print(cv_score)
print('평균 :', cv_score.mean())
print('표준편차 :', cv_score.std())
```

#### knn_k, k-fold_k를 for문에 넣어주기

```python
result = []
my_k = []
my_cv = []

for i in range(3,22,2):
    for c in range(5,11,1):
        model = KNeighborsRegressor(n_neighbors=i)
        cv_score = cross_val_score(model,x_train,y_train,cv=c)
        
        result.append(cv_score.mean())
        my_k.append(i)
        my_cv.append(c)
        
score = pd.DataFrame()
score['knn_k'] = my_k
score['cv'] = my_cv
score['result'] = result

score.sort_values('result', ascending=False, inplace=True)
score.head(10)
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor       # y가 수치형 이기 때문에
from sklearn.model_selection import cross_val_score

model = DecisionTreeRegressor()

cv_score = cross_val_score(model,x_train,y_train,cv=10)

print(cv_score)
print('평균 :', cv_score.mean())
print('표준편차 :', cv_score.std())
```
